```python
def cate(cols):
    GS = cols[0]
    cat = cols[1]
    if pd.isnull(GS):

        if cat == 0:
            return 3

        elif cat == 1:
            return 8
        
        elif cat == 2:
            return 20
        
        elif cat == 3:
            return 33.5
        
        elif cat == 4:
            return 39
        
        elif cat == 5:
            return 52
        
        elif cat == 6:
            return 61
        
        elif cat == 7:
            return 58
        
        elif cat == 8:
            return 56.5
        
        elif cat == 9:
            return 73.5
        
        elif cat == 10:
            return 74
        
        elif cat == 11:
            return 68
        
        elif cat == 12:
            return 88
        
        elif cat == 13:
            return 102
        
        elif cat == 14:
            return 106

        else:
            return 82

    else:
        return GS
```

```python
X = df.dropna().drop(['playerName','NBAPTS', 'NBATRB', 'NBAAST', 'NBASTL', 'NBABLK'], axis=1)
y =  df.dropna()['NBAPTS']
col = list(X.columns)
score_list=[]
high_score=0
nof=0           
for n in range(1,len(col)):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 101)
    model = LinearRegression()
    rfe = RFE(model,n)
    X_train_rfe = rfe.fit_transform(X_train,y_train)
    X_test_rfe = rfe.transform(X_test)
    model.fit(X_train_rfe,y_train)
    score = model.score(X_test_rfe,y_test)
    score_list.append(score)
    if(score>high_score):
        high_score = score
        nof = n
        print("Optimum number of features: %d" %nof)
print("Score with %d features: %f" % (nof, high_score))
```
